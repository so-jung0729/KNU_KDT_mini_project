```python
# 그래프
import matplotlib.pyplot as plt
```

```python
# 출생아수 파일 불러오기
FILE1='../project_01/출생아수__합계출산율__자연증가_등_20230106005447.csv'
Birth=pd.read_csv(FILE1,encoding='cp949')
```

```python
Birth.head(), Birth.info()
```

```python
Birth
```

```python
# 혼인건수 파일 불러오기
FILE2='../project_01/혼인건수__조혼인율_20230106005343.csv'
Marriage=pd.read_csv(FILE2,encoding='cp949')
```

```python
Marriage
```

```python
# 자료 합치기
B_M=pd.concat([Birth,Marriage])
B_M
```

```python
# 출산, 결혼 필요한 값만 저장
BM_select=B_M.loc[0]    # 필요한 값 둘다 인덱스가 0
BM_index=BM_select.set_index(['기본항목별'])
```

```python
BM_index.info()
```

```python
BM_index
```

```python
# GNI 파일 불러오기
FILE3='../project_01/GNI_excel.xlsx'
GNI=pd.read_excel(FILE3,header=2,index_col=1)
GNI
```

```python
GNI_drop=GNI.drop('Unnamed: 0',axis=1)
GNI_drop
```

```python
GNI_select=GNI_drop.loc['실질/원화']
GNI_select, GNI_select.info(), type(GNI_select)
```

```python
# merge하기 위해서 frame으로 바꿈.
GNIDF=GNI_select.to_frame().reset_index()
GNIDF
```

```python
BM_T=BM_index.T.reset_index()
BM_T
```

```python
BMG=pd.merge(BM_T,GNIDF)
BMG
```

```python
BMG=BMG.rename(columns={'index':'year','출생아수(명)':'birth','혼인건수(건)':'marriage','실질/원화':'GNI'})
BMG
```

```python
# GNI값 실수로 바꾸기 위해서 ',' 제거
BMG.GNI=BMG.GNI.apply(lambda x: float(x.replace(',', '')))
```

```python
BMG.info()
```

```python
fig, ax1 = plt.subplots(figsize = (20, 10))
ax2 = ax1.twinx()

ax1.plot(BMG.year,BMG.GNI, 'ro-', label = 'GNI')
ax2.plot(BMG.year,BMG.marriage, 'bo-', label ='marriage')
ax2.plot(BMG.year,BMG.birth, 'yo-', label = 'birth')
```

```python
ax1.set_ylabel('GNI')
```

```python
fig
```

```python
# Birth 산포도 
plt.scatter(BMG.GNI,BMG.birth,color='orange')
```

```python
# Marriage 산포도 
plt.scatter(BMG.GNI,BMG.marriage, color='red')
```

```python
cor=BMG.corr(method='pearson')
cor
```

출생아수, 혼인건수 모두 대략 -0.8 정도로 GNI와 강한 음의 상관관계가 있다고 볼 수 있습니다.
즉, GNI가 높아질 수록 출생아수와 혼인건수가 낮아지는 것을 확인 할 수 있습니다.
그에 대한 원인으로는 영유아 사망률 감소의 영향, 가족 구성원 간 자원흐름 변화, 질과 양의 대체 등으로 추측할 수 있습니다.
